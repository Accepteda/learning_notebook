```python
import pandas as pd
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
emaotai = pd.read_csv(r'emaotai.csv')
emaotai['时间'] = emaotai['时间'].str[:10]
```

```python
emaotai
```

```python
emaotai_0808 = emaotai[emaotai['时间']== '2017-08-08']
emaotai_0816 = emaotai[emaotai['时间']== '2017-08-16']
```

```python
emaotai_0808['销售额'] = emaotai_0808['价格']*emaotai_0808['已售出量']
emaotai_0816['销售额'] = emaotai_0816['价格']*emaotai_0816['已售出量']
```

```python
maotai_price_0816 = emaotai_0816.sort_values(by='价格', ascending=False)
maotai_sales_count_0816 = emaotai_0816.sort_values(by='已售出量', ascending=False)
maotai_sales_income_0816 = emaotai_0816.sort_values(by='销售额', ascending=False)
maotai_view_count_0816 = emaotai_0816.sort_values(by='浏览量', ascending=False)
```
