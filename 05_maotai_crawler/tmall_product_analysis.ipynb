```python
#茅台醇属于技术开发公司，悠蜜为生态农业公司，葡萄酒为葡萄酒公司
maotai_labels = ['王子酒', '悠蜜', '汉酱酒', '葡萄酒', '迎宾', '赖茅', '茅台醇', '贵州大曲', '仁酒']
maotai_label_list = []
for i in maotai_tmall['商品名']:
    for j in maotai_labels:
        if j in i:
            maotai_label_list.append(j)
            break
    else:
        maotai_label_list.append('伊利')
maotai_tmall['品牌'] = maotai_label_list
```

```python
maotai_tmall['销售额'] = maotai_tmall['价格'] * maotai_tmall['销量']
```

```python
mtmall = maotai_tmall[(maotai_tmall['品牌'] != '悠蜜') & (maotai_tmall['品牌'] != '葡萄酒') & (maotai_tmall['品牌'] != '贵州醇')]
```

```python
mtmall
```

```python
maotai_0811 = mtmall[mtmall['时间'] == '2017-8-11']
maotai_0816 = mtmall[mtmall['时间'] == '2017-8-16']
maotai_0821 = mtmall[mtmall['时间'] == '2017-08-21']
```
