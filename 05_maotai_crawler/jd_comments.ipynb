```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
#导入结巴分词库(分词) 
import jieba as jb 
#导入结巴分词(关键词提取) 
import jieba.analyse
from wordcloud import WordCloud
%matplotlib inline
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
comment_urls = ['https://club.jd.com/comment/productPageComments.action?productId=251837&score=0&sortType=6&page=%d&pageSize=10' % (i) for i in range(100)]
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0'
}
```

```python
def get_page(url):
    try:
        r = requests.get(url, headers=headers)
        r.raise_for_status
        return r.json()
    except:
        print(url)
        return
```

```python
comments = []
for url in tqdm(comment_urls):
    data = get_page(url)
    try:
        comments += data['comments']
    except:
        continue
```

```python
len(comments)
```

```python
nickname = [comment['nickname'] for comment in comments]
content = [comment['content'] for comment in comments]
productColor = [comment['productColor'] for comment in comments]
referenceName = [comment['referenceName'] for comment in comments]
creationTime = [comment['creationTime'] for comment in comments]
days = [comment['days'] for comment in comments]
referenceTime = [comment['referenceTime'] for comment in comments]
score = [comment['score'] for comment in comments]
replyCount = [comment['replyCount'] for comment in comments]
usefulVoteCount = [comment['usefulVoteCount'] for comment in comments]
uselessVoteCount = [comment['uselessVoteCount'] for comment in comments]
isMobile = [comment['isMobile'] for comment in comments]
userClient = [comment['userClient'] for comment in comments]
userClientShow = [comment['userClientShow'] for comment in comments]
userLevelName = [comment['userLevelName'] for comment in comments]
```

```python
table=pd.DataFrame({'productColor':productColor,
                    'referenceName':referenceName,
                    'creationTime':creationTime,
                    'referenceTime':referenceTime,
                    'days':days,
                    'nickname':nickname,
                    'content':content,
                    'productColor':productColor,
                    'isMobile':isMobile,
                    'userClient':userClient,
                    'userClientShow':userClientShow,
                    'userLevelName':userLevelName,
                    'replyCount':replyCount,
                    'usefulVoteCount':usefulVoteCount,
                    'uselessVoteCount':uselessVoteCount,
                    'score':score})
```

```python
date = time.strftime('%Y%m%d',time.localtime(time.time()))
table.to_csv(r'data/jd_comments%s.csv' % (date))
```

```python
table['referenceTime']=pd.to_datetime(table['referenceTime'])
table['creationTime']=pd.to_datetime(table['creationTime'])
table = table.set_index('referenceTime')
table['days']=table['days'].astype(np.int64)
hour=[] 
for h in referenceTime: 
    date=h[10:13] 
    hour.append(date)
table['hour'] = hour
month=[] 
for m in referenceTime: 
    date=m[5:7] 
    month.append(date)
table['month'] = month
```

```python
product_name = []
for i in table.productColor:
    if '飞天53度 500ml 单瓶' == i or '飞天53度/单瓶' == i or i == '':
        product_name.append('500ml')
    elif i == '飞天53度 500ml 整箱' or i == '飞天53度/整箱':
        product_name.append('500ml整箱')
    elif '茅台 飞天 53度 400ml  单瓶' == i or '茅台 飞天53度 400ml' == i:
        product_name.append('400ml')
    elif '茅台 飞天 53度  750ml' == i:
        product_name.append('750ml')
    elif '飞天茅台 53度 1.5L' == i:
        product_name.append('1.5L')
    elif '飞天茅台 53度 3L' == i:
        product_name.append('3L')
    else:
        product_name.append('6L')
table['product_name'] = product_name
```

```python
ClientShow = []
for i in table.userClientShow:
    if i == '':
        i = 'PC'
    else:
        i = i[2:]
    ClientShow.append(i)
table.userClientShow = ClientShow
```
